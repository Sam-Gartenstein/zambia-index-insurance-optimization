# Zambia Insurance Optimization

The goal of this project is to reduce basis risk in weather insurance contracts, building upon [earlier work](https://github.com/Sam-Gartenstein/zambia-drought-analysis) where we introduced the Payout Balance Index (PBI). Rather than relying on K-means clustering, we calculate the PBI using Zambiaâ€™s two administrative levels along with camps developed by (name), and estimate the monetary value of insurance forgone due to mismatches between payouts and actual conditions. In addition, we consider the perspective of a decision-maker, who must weigh the administrative burden of monitoring each zone. Framing the problem as a constrained optimization, we seek to minimize basis risk (equivalently, maximize the number of farmers protected) while simultaneously reducing administrative complexity.

## Mapping Missing Pixel Values  

The pixel data collected from GEE was extracted using FAO GAUL 2015 boundary definitions. Since the most up-to-date boundaries differ slightly in shape, some pixels may fall outside the newer administrative limits. Below, I analyze missing pixel data for each district and camp.

```python
def plot_missing_points(points_gdf, boundary_gdf, label_col, 
                        title=None, marker_size=12,
                        boundary_kwargs=None, point_kwargs=None):
    """
    Plot features in `points_gdf` where `label_col` is missing, overlaid on `boundary_gdf`.

    - Auto-aligns CRS by reprojecting points to boundary CRS if needed (does NOT force EPSG:4326).
    - Returns (missing_points_gdf, ax) for further use.
    """
    boundary_kwargs = boundary_kwargs or {"color": "black", "linewidth": 1}
    point_kwargs = point_kwargs or {"color": "red", "markersize": marker_size, "label": f"Missing {label_col}"}

    # 1) Filter rows with missing label
    missing = points_gdf[points_gdf[label_col].isna()].copy()

    # 2) Align CRS (only if different)
    if boundary_gdf.crs is None and missing.crs is None:
        # No CRS info anywhereâ€”plot will still work, but spatial alignment isn't guaranteed
        pass
    elif boundary_gdf.crs is not None and (missing.crs is None or missing.crs != boundary_gdf.crs):
        missing = missing.to_crs(boundary_gdf.crs)

    # 3) Plot
    fig, ax = plt.subplots(figsize=(10, 10))
    boundary_gdf.boundary.plot(ax=ax, **boundary_kwargs)
    if len(missing):
        missing.plot(ax=ax, **point_kwargs)
        ax.legend()
    ax.set_title(title or f"Locations with Missing '{label_col}'", fontsize=14)
    plt.show()

    return missing, ax
```

### Spatial Aggregation to Administrative District  

We now aggregate pixel-level values to the district level by taking the mean of all pixels within each district for every year. This approach mirrors Google Earth Engineâ€™s method for spatial aggregation when using the API, ensuring consistency between pixel-level data and administrative boundaries.  


```python
'''
Move to script
'''

def compute_mean_precip(df, group_cols, value_col="precipitation"):
    """
    Groups the DataFrame by specified columns and computes the mean of the value column.

    Parameters:
    - df (DataFrame): Input DataFrame.
    - group_cols (list): List of columns to group by.
    - value_col (str): Name of the column to average.

    Returns:
    - DataFrame with group means.
    """
    return df.groupby(group_cols)[value_col].mean().reset_index()
```

## Payout Balance Metric

To evaluate how well the clustering minimizes **basis risk**, we first compute the **absolute deviation** from a 50% payout rate for each cluster-year pair:

$$
D_t^k = |P_t^k - 50|
$$

Where:

- $P_t^k$ = Percentage of pixels receiving a payout in **cluster** $k$ during **year** $t$  
- $D_t^k$ = Absolute deviation from the 50% balance point â€” **higher values indicate greater basis risk**

---

### Payout Balance Index (PBI)

We then transform this deviation into a **Payout Balance Index (PBI)** scaled between 0 and 100:

$$
\text{PBI}_{t,k} = \left(1 - \frac{D_t^k}{50} \right) \times 100
$$

- A value of **100** means perfect balance: all or none of the farmers in the cluster received payouts  
- A value of **0** indicates a 50/50 split â€” the **worst-case scenario** for basis risk

---

### Overall Clustering Quality

To assess clustering quality across the entire dataset, we compute the **average PBI**:

$$
\text{PBI}_{\text{avg}} = \frac{1}{N} \sum_{t,k} \text{PBI}_{t,k}
$$

Where $N$ is the total number of (year Ã— cluster) combinations.

---

### Basis Risk Interpretation

- A **PBI close to 100** reflects **low basis risk**: clusters are internally consistent â€” farmers in the same area experience similar insurance outcomes  
- A **PBI near 0** reflects **high basis risk**: payout decisions vary significantly within clusters, indicating poor targeting

This index provides a **clear, interpretable, and policy-relevant** measure of how well your spatial clustering aligns with actual drought impact.

## Forgone Payouts  

The next step is to calculate **forgone payouts**, which arise from **basis risk**.  
The **Payout Balance Index (PBI)** is designed to capture both dimensions of basis risk:  

- **Overpayments** â€” payouts made when they should not have been, and  
- **Underpayments** â€” payouts missed when they should have been triggered.  

Thus, the share of payouts not aligned with actual drought outcomes is represented by:  

$$
1 - \text{PBI}
$$  

The **forgone payout** is then computed as:  

$$
\text{Forgone Payout} =
(\text{Payout Value}) \times (1 - \text{PBI}) \times (\text{Number of Pixels Per Zone})
$$  

### Parameters and Assumptions  

- **Payout Value**: assumed to be **1600 Kwacha** (â‰ˆ \$68).  
  - This figure is derived by dividing the total payout of **800 million Kwacha** by **500,071 smallholder farmers** who received payments for climate-related losses ([Pula Advisors, 2022](https://www.pula-advisors.com/post/zambia-s-largest-crop-index-insurance-payout-marks-historic-milestone-for-smallholder-farmers)).  

- **Number of Pixels**:  
  - ADM1 and ADM2 â†’ **25,022 pixels**  
  - ADM3 (camps) â†’ **25,054 pixels**  

- **Key Assumption**: each pixel represents an individual farmer, and farmers are assumed to be evenly distributed geographically.  



```python
# Define base payout and total number of pixels at each administrative level

payout_value = 1600        # Insurance payout value (currency units, e.g., ZMW)

admin_1_pixels = 25022     # Total number of pixels aggregated at Admin Level 1 (provinces/districts)
admin_2_pixels = 25022     # Total number of pixels aggregated at Admin Level 2 (sub-districts/wards)
admin_3_pixels = 25054     # Total number of pixels aggregated at Admin Level 3 (camps)
```

### Administrative 1

```python
# Administrative 1 forgone payments
admin_1_fp = (payout_value) * (1 - (payout_score_admin_1/100)) * (admin_1_pixels) 
print(f"Administrative 1 forgone payments: {admin_1_fp:.2f} kwacha")
```

### Administrative 2

```python
# Administrative 2 forgone payments
admin_2_fp = (payout_value) * (1 - (payout_score_admin_2/100)) * (admin_2_pixels) 
print(f"Administrative 2 forgone payments: {admin_2_fp:.2f} kwacha")
```

### Administrative 3

```python
# Administrative 3 forgone payments
admin_3_fp = (payout_value) * (1 - (payout_score_admin_3/100)) * (admin_3_pixels) 
print(f"Administrative 3 forgone payments: {admin_3_fp:.2f} kwacha")
```

## Marginal Benefit per Zone Change

Increasing the number of zones reduces forgone payments, but this does not automatically imply greater implementation efficiency. As zone counts grow, diminishing returns may ariseâ€”each added zone tends to yield a smaller improvement.

To quantify the marginal effect of moving from zone $i$ to zone $j$ ($i \to j$), define

$$
m_{i \to j} = \frac{\lvert FP_j - FP_i \rvert}{Z_j - Z_i},
$$

where $FP_k$ is the forgone payment at level $k$ and $Z_k$ is the number of zones at that level. The quantity $m_{i \to j}$ measures the absolute change in forgone payments per additional zone (ZMW per zone). Remove the absolute value if a signed measure is preferred.


```python
"""
Increases in number of zones (Î” zones = destination âˆ’ source)
"""

# Admin 1 (provinces) â†’ K-means clusters
inc_1k = num_clusters - num_provinces   # additional zones when moving from provinces to clusters

# K-means clusters â†’ Admin 2 (districts)
inc_k2 = num_districts - num_clusters   # additional zones when moving from clusters to districts

# Admin 2 (districts) â†’ Camps
inc_2c = num_unique_camps - num_districts      # additional zones when moving from districts to camps
```

```python
"""
Changes in forgone payments (Î”FP = destination âˆ’ source; units: ZMW)
Positive Î”FP â†’ increase in forgone payments; negative Î”FP â†’ reduction.
"""

# Admin 1 (provinces) â†’ K-means clusters
d_fp_1k = clusters_fp - admin_1_fp   # change when moving from provinces to clustered zones

# K-means clusters â†’ Admin 2 (districts)
d_fp_k2 = admin_2_fp - clusters_fp   # change when moving from clusters to districts

# Admin 2 (districts) â†’ Camps
d_fp_2c = admin_3_fp - admin_2_fp    # change when moving from districts to camps
```

```python
"""
Marginal absolute change in forgone payments per added zone
Formula: m_{iâ†’j} = |Î”FP_{iâ†’j}| / Î”Z_{iâ†’j}
Units: ZMW per added zone
Note: assumes Î”Z > 0 for each transition.
"""

# Admin 1 (provinces) â†’ K-means clusters
marg_1k = abs(d_fp_1k) / inc_1k  

# K-means clusters â†’ Admin 2 (districts)
marg_k2 = abs(d_fp_k2) / inc_k2  

# Admin 2 (districts) â†’ Camps
marg_2c = abs(d_fp_2c) / inc_2c  
```

```python
print("Admin 1 to Clusters")
print("Increase in zones:", inc_1k)
print("Absolute change in forgone payments in million zambia kwacha:", round(abs(d_fp_1k)/1e6, 4))
print("Absolute change per added zone in zambia kwacha:", round(marg_1k, 2))

print("\nClusters to Admin 2")
print("Increase in zones:", inc_k2)
print("Absolute change in forgone payments in million zambia kwacha:", round(abs(d_fp_k2)/1e6, 4))
print("Absolute change per added zone in zambia kwacha:", round(marg_k2, 2))

print("\nAdmin 2 to Camps")
print("Increase in zones:", inc_2c)
print("Absolute change in forgone payments in million zambia kwacha:", round(abs(d_fp_2c)/1e6, 4))
print("Absolute change per added zone in zambia kwacha:", round(marg_2c, 2))
```

```python
vals = [round(abs(d_fp_1k)/1e6, 4), round(abs(d_fp_k2)/1e6, 4), round(abs(d_fp_2c)/1e6, 4)]
fig, ax = benefit_changes(
    vals,
    title="Total Benefit for Zone Change",
    annotate_offsets=[(0, 10), (17, -7), (-17, -4)],  # per-point placement
    annotate_ha="center",
    annotate_va="bottom",
 )
plt.show()
```

```python
vals = [marg_1k, marg_k2, marg_2c]
fig, ax = benefit_changes(
    vals,
    title="Marginal Benefit per Zone Change",
    annotate_offsets=[(35, -6), (-30, -6), (3, 8)],  # per-point placement
    annotate_ha="center",
    annotate_va="bottom",
 )
plt.show()
```
